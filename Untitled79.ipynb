```python
# Step 1: Setup
!pip install nltk
import os
import nltk
nltk.download('punkt')
import zipfile
import urllib.request
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import random
import string

# Create working dir
dataset_dir = '/content/flickr8k'
os.makedirs(dataset_dir, exist_ok=True)

# Download Flickr8k images and captions
image_url = 'https://github.com/jbrownlee/Datasets/releases/download/Flickr8k/Flickr8k_Dataset.zip'
caption_url = 'https://github.com/jbrownlee/Datasets/releases/download/Flickr8k/Flickr8k_text.zip'

# Helper function
def download_and_extract(url, extract_to):
    zip_path = os.path.join(extract_to, os.path.basename(url))
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    os.remove(zip_path)

# Download images and captions
download_and_extract(image_url, dataset_dir)
download_and_extract(caption_url, dataset_dir)
```

```python
# Read the caption file
caption_path = os.path.join(dataset_dir, 'Flickr8k.token.txt')

# Load captions into a dictionary
def load_captions(caption_file):
    captions_dict = {}
    with open(caption_file, 'r') as f:
        for line in f:
            tokens = line.strip().split('\t')
            img_id, caption = tokens[0].split('#')[0], tokens[1]
            caption = caption.lower().translate(str.maketrans('', '', string.punctuation))
            if img_id not in captions_dict:
                captions_dict[img_id] = []
            captions_dict[img_id].append(caption)
    return captions_dict

captions_dict = load_captions(caption_path)

# Show how many images and captions
print(f"Total images: {len(captions_dict)}")
```

```python
# Visualize a few samples
images_path = os.path.join(dataset_dir, 'Flicker8k_Dataset')
sample_images = random.sample(list(captions_dict.keys()), 3)

for img_id in sample_images:
    img_path = os.path.join(images_path, img_id)
    img = Image.open(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title("\n".join(captions_dict[img_id][:2]), fontsize=10)
    plt.show()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Prepare all captions in a list
all_captions = []
image_ids = []

for img_id, caps in captions_dict.items():
    for c in caps:
        all_captions.append('startseq ' + c + ' endseq')
        image_ids.append(img_id)

print(f"Total processed captions: {len(all_captions)}")

# Tokenize the captions
tokenizer = Tokenizer(num_words=5000, oov_token='unk')
tokenizer.fit_on_texts(all_captions)
vocab_size = len(tokenizer.word_index) + 1
print(f"Vocabulary size: {vocab_size}")

# Save tokenizer for future use
import pickle
with open("tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)
```

```python
import tensorflow as tf
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Model
from tqdm import tqdm

# Load pretrained model
base_model = InceptionV3(weights='imagenet')
model = Model(base_model.input, base_model.layers[-2].output)

def extract_features(directory, limit=1000):
    features = {}
    model_input_shape = (299, 299)
    count = 0
    for img_name in tqdm(os.listdir(directory)):
        if not img_name.lower().endswith('.jpg'):
            continue
        if count >= limit:
            break
        try:
            img_path = os.path.join(directory, img_name)
            image = load_img(img_path, target_size=model_input_shape)
            image = img_to_array(image)
            image = np.expand_dims(image, axis=0)
            image = preprocess_input(image)
            feature = model.predict(image, verbose=0)
            features[img_name] = feature
            count += 1
        except:
            continue
    return features

# Run with a limit (e.g., 1000 images)
features = extract_features(images_path, limit=1000)

# Save the reduced feature set
with open("image_features.pkl", "wb") as f:
    pickle.dump(features, f)

print(f" Extracted features for {len(features)} images")
```

```python
# Load saved tokenizer
with open("tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

# Get max caption length from raw captions, not tokenized ones
max_length = max(len(c.split()) for c in all_captions)
print(f"Maximum caption length: {max_length}")

# Prepare sequences for model
def create_sequences(tokenizer, max_length, features, captions_dict):
    X1, X2, y = [], [], []
    for img_id, captions in captions_dict.items():
        if img_id not in features:
            continue
        for cap in captions:
            cap = 'startseq ' + cap + ' endseq'
            seq = tokenizer.texts_to_sequences([cap])[0]
            for i in range(1, len(seq)):
                in_seq, out_seq = seq[:i], seq[i]
                in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
                out_seq = tf.keras.utils.to_categorical(out_seq, num_classes=vocab_size)
                X1.append(features[img_id][0])
                X2.append(in_seq)
                y.append(out_seq)
    return np.array(X1), np.array(X2), np.array(y)

X1, X2, y = create_sequences(tokenizer, max_length, features, captions_dict)
print(f"X1 shape: {X1.shape}, X2 shape: {X2.shape}, y shape: {y.shape}")
```

```python
# Prepare a small dataset for memory-safe training
X1, X2, y = [], [], []

count = 0
for img_id, captions in captions_dict.items():
    if img_id not in features:
        continue
    for cap in captions[:1]:  # just 1 caption per image (simplify!)
        cap = 'startseq ' + cap + ' endseq'
        seq = tokenizer.texts_to_sequences([cap])[0]
        for i in range(1, len(seq)):
            in_seq, out_seq = seq[:i], seq[i]
            in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
            out_seq = tf.keras.utils.to_categorical(out_seq, num_classes=vocab_size)
            X1.append(features[img_id][0])
            X2.append(in_seq)
            y.append(out_seq)
    count += 1
    if count == 100:  # train on 100 images max (you can increase later)
        break

X1 = np.array(X1)
X2 = np.array(X2)
y = np.array(y)

print(f"Shapes - X1: {X1.shape}, X2: {X2.shape}, y: {y.shape}")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add

inputs1 = Input(shape=(2048,))  #  FIXED this line
fe1 = Dropout(0.5)(inputs1)
fe2 = Dense(256, activation='relu')(fe1)

inputs2 = Input(shape=(max_length,))
se1 = Embedding(vocab_size, 256)(inputs2)  # mask_zero removed
se2 = Dropout(0.5)(se1)
se3 = LSTM(256)(se2)

decoder1 = add([fe2, se3])
decoder2 = Dense(256, activation='relu')(decoder1)
outputs = Dense(vocab_size, activation='softmax')(decoder2)

model = Model(inputs=[inputs1, inputs2], outputs=outputs)
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.summary()
```

```python
from tensorflow.keras.utils import Sequence

class CaptionDataGenerator(Sequence):
    def __init__(self, captions_dict, features, tokenizer, max_length, vocab_size, batch_size=32):
        self.captions_dict = captions_dict
        self.features = features
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.vocab_size = vocab_size
        self.batch_size = batch_size
        self.image_ids = list(captions_dict.keys())
        self.data = self._create_data()

    def _create_data(self):
        # Flatten all (image_id, caption) pairs
        data = []
        for img_id, captions in self.captions_dict.items():
            if img_id not in self.features:
                continue
            for cap in captions:
                data.append((img_id, 'startseq ' + cap + ' endseq'))
        return data

    def __len__(self):
        return int(np.ceil(len(self.data) / self.batch_size))

    def __getitem__(self, idx):
        X1, X2, y = [], [], []
        batch = self.data[idx * self.batch_size:(idx + 1) * self.batch_size]
        for img_id, caption in batch:
            seq = self.tokenizer.texts_to_sequences([caption])[0]
            for i in range(1, len(seq)):
                in_seq, out_seq = seq[:i], seq[i]
                in_seq = pad_sequences([in_seq], maxlen=self.max_length)[0]
                out_seq = tf.keras.utils.to_categorical(out_seq, num_classes=self.vocab_size)
                X1.append(self.features[img_id][0])
                X2.append(in_seq)
                y.append(out_seq)
        return [np.array(X1), np.array(X2)], np.array(y)
```

```python
model.fit([X1, X2], y, epochs=10, batch_size=32)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

def generate_caption(model, tokenizer, photo_feature, max_length):
    in_text = 'startseq'
    for _ in range(max_length):
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = model.predict([photo_feature, sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = tokenizer.index_word.get(yhat)
        if word is None:
            break
        in_text += ' ' + word
        if word == 'endseq':
            break
    final_caption = in_text.replace('startseq', '').replace('endseq', '').strip()
    return final_caption
```
